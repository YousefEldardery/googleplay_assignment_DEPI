fix size

```python
df['Size'].value_counts()
```

```python
# convert the kilobyte size to megabyte and the Varies with device to np.nan
def fix_size(s):
    try:
        if s.endswith('M'):
            return float(s[:-1])
        elif s.endswith('k'):
            return float(s[:-1]) / 1024
        elif s == '1,000+':
            return 1.0
        else:
            return np.nan
    except:
        return np.nan
        
```

```python
#apply the size function 
df['Size'] = df['Size'].apply(fix_size)
```

fix price

```python
df['Price'].value_counts()
```

```python
def fix_price(p):
    try:
        if p.startswith('$'):
            return float(p[1:])
        else:
            return 0.0
    except:
        return np.nan
```

```python
df['Price']=df['Price'].apply(fix_price)
```

```python
from sklearn.impute import KNNImputer

# Columns to impute
columns_to_impute = ['Rating', 'Size','Minimum_version']  # Add the column names you want to impute

# Initialize KNNImputer
imputer = KNNImputer(n_neighbors=3)

# Select numeric and categorical columns
Num_df = df.select_dtypes(include="number")
Cat_df = df.select_dtypes(include="object")

# Impute missing values only for specified columns
Num_df[columns_to_impute] = imputer.fit_transform(Num_df[columns_to_impute])

# Concatenate numeric and categorical data
df = pd.concat([Num_df, Cat_df], axis=1)
```
